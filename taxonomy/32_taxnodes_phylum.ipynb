```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
file_path = "../32_pseudomonadota/phylum-phylumId_phylumName_superkingdom_superkingdomName_count_nSpecies.tsv"
df = pd.read_csv(file_path, sep="\t")

# Ensure correct column names (assuming the order based on your description)
df.columns = ["phylumId", "phylumName", "superkingdom", "superkingdomName", "count", "nSpecies"]

# Get unique superkingdoms
superkingdoms = df["superkingdomName"].unique()

# Create a figure with subplots
fig, axes = plt.subplots(1, len(superkingdoms), figsize=(5 * len(superkingdoms), 5), sharey=False)

# Plot each superkingdom in a separate subplot
for ax, sk in zip(axes, superkingdoms):
    subset = df[df["superkingdomName"] == sk]
    
    ax.scatter(subset["nSpecies"], subset["count"], alpha=0.7)
    ax.set_title(sk)
    ax.set_xlabel("#species")
    ax.set_ylabel("#clu")
    ax.set_xscale('log')
    ax.set_yscale('log')
    # ax.set_xlim(0-, max(subset['nSpecies']))
    # ax.set_ylim(0, max(subset['count']))

# Adjust layout and show plot
plt.tight_layout()
plt.savefig('output/phylum_proteinCount_nSpecies.svg')
```
