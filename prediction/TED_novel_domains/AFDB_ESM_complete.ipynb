```python
import pandas as pd

df = pd.read_csv("../19_pred_novel_domains/TED_AFDB_ESM_domain-entryId_AFDBavgPlddt_ESMavgPlddt_lddt.tsv", \
  delimiter="\t", names="entryId_AFDBavgPlddt_ESMavgPlddt_lddt".split("_"), header=None)

df.head()
```

```python
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec

# Create a figure with a GridSpec layout
fig = plt.figure(figsize=(8, 8))
gs = GridSpec(4, 4, hspace=0.4, wspace=0.4)

# Scatter plot (central panel)
ax_scatter = fig.add_subplot(gs[1:4, 0:3])
ax_scatter.scatter(df["ESMavgPlddt"], df["AFDBavgPlddt"], alpha=0.1)
ax_scatter.set_xlim([df["ESMavgPlddt"].min() - 5, 100])
ax_scatter.set_ylim([df["ESMavgPlddt"].min() - 5, 100])
ax_scatter.set_xlabel("ESMavgPlddt")
ax_scatter.set_ylabel("AFDBavgPlddt")

# Histogram for X-axis (top panel)
ax_hist_x = fig.add_subplot(gs[0, 0:3], sharex=ax_scatter)
ax_hist_x.hist(df["ESMavgPlddt"], bins=20, alpha=0.6, color="blue")
ax_hist_x.set_ylabel("Frequency")
ax_hist_x.tick_params(axis="x", labelbottom=False)  # Hide x-axis labels

# Histogram for Y-axis (right panel)
ax_hist_y = fig.add_subplot(gs[1:4, 3], sharey=ax_scatter)
ax_hist_y.hist(df["AFDBavgPlddt"], bins=20, orientation="horizontal", alpha=0.6, color="red")
ax_hist_y.set_xlabel("Frequency")
ax_hist_y.tick_params(axis="y", labelleft=False)  # Hide y-axis labels

# Title for the whole figure
fig.suptitle("Scatter Plot with Marginal Histograms")

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming df is your DataFrame with two numeric columns, e.g., x and y
x = df["ESMavgPlddt"]
y = df["AFDBavgPlddt"]

# Initialize a JointGrid
g = sns.JointGrid(x=x, y=y, space=0, height=6)

# KDE Contours with scatter overlay
# g.plot_joint(plt.scatter, s=10, alpha=0.2, color="darkblue")  # Scatter Plot
g.plot_joint(sns.kdeplot, 
             cmap="mako",  # Background gradient
             levels=20, 
             fill=True,
             linewidths=1.5, 
             color="black")   # Darker contour lines

# Marginal histograms
g.plot_marginals(sns.histplot, bins=20, kde=False, color="navy")

# Set axis limits
g.ax_joint.set_xlim([df["ESMavgPlddt"].min() - 5, 105])
g.ax_joint.set_ylim([df["ESMavgPlddt"].min() - 5, 105])

# Update marginal axes to reflect the same limits
g.ax_marg_x.set_xlim([df["ESMavgPlddt"].min() - 5, 105])
g.ax_marg_y.set_ylim([df["ESMavgPlddt"].min() - 5, 105])

# Final touches
g.figure.suptitle("2D Density Plot with Marginal Histograms", y=1.05)
g.ax_joint.set_xlabel("ESMavgPlddt")
g.ax_joint.set_ylabel("AFDBavgPlddt")

plt.savefig('output/AFDB-ESM.svg')
```
